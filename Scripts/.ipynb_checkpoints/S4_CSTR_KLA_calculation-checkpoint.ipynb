```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Needed to calculat kLa in the aerobic reactor for heterotrophic fermentation of acetate

From book: Enes Kadic, Theodore J. Heindel, 2014, An Introduction to Bioreactor Hydrodynamics and Gas‐Liquid Mass Transfer

Chapter 6: Stirre-tank Bioreactors

Equation 6.10, Table 6.3

Equations:

<p> Mass transfer coefficient: </p> 
<div style="text-align: center">  $ k_La = C * (N/N_{CD})^A * v^B_{gs}$ </div>

<p>With:</p>

 - A, B, and C being fitted constants
 - N being the impeller speed, and N_CD the impeller speed at which a transition in flow regime occurs from loaded to completely dispersed [rev/s]
 - v_GS the Superficial gas velocity [m∕s]
   

<p> Superficial gas velocity: </p> 
<div style="text-align: center">  $ v^c_{GS} = F_av/A/3600 $ </div>

<p>With:</p>

 - A being the reactor cross-sectional area [m^2]
 
<p>Assuming that the impeller is operated exactly at the speed of N_CD, we have:</p>
<div style="text-align: center">  $ k_La = C * v^B_{gs}$ </div>

With the fitted constants from Table 6.3, 

Comparisson for various sets of fitted constants:

```python
def kLa_Chandrasekharan(v_gas):
    return 2.7*v_gas*0.96
def kLa_Calderbank(v_gas):
    return 2.8*v_gas*0.97
def kLa_Hickman(v_gas):
    return 4.3*v_gas*1.04
def kLa_Linek(v_gas):
    return 5.2*v_gas*1.09
def kLa_Smith1977(v_gas):
    return 2.4*v_gas*0.96
def kLa_Smith1991(v_gas):
    return 6.5*v_gas*1.12
def kLa_SmithWarmoeskerken(v_gas):
    return 12.6*v_gas*1.27
def kLa_WhittonNienow(v_gas):
    return 3.5*v_gas*1
def kLa_Yawalkar(v_gas):
    return 3.4*v_gas*1
def kLa_Zhu(v_gas):
    return 3.3*v_gas*0.97
```

```python
v_gas = np.arange(0, 20, 0.1)

fig = plt.figure(figsize=(10,5))
plt.title(f"Volumetric mass transfer coefficient KLa by gas velocity (log)")
plt.plot(v_gas, kLa_Chandrasekharan(v_gas))
plt.plot(v_gas, kLa_Calderbank(v_gas))
plt.plot(v_gas, kLa_Hickman(v_gas))
plt.plot(v_gas, kLa_Linek(v_gas))
plt.plot(v_gas, kLa_Smith1977(v_gas))
plt.plot(v_gas, kLa_Smith1991(v_gas))
plt.plot(v_gas, kLa_SmithWarmoeskerken(v_gas))
plt.plot(v_gas, kLa_WhittonNienow(v_gas))
plt.plot(v_gas, kLa_Yawalkar(v_gas))
plt.plot(v_gas, kLa_Zhu(v_gas))
plt.legend(["Chandrasekharan","Calderbank", 'Hickman','Linek','Smith 1977', 'Smith 1991', 'Smith & Warmoeskerken', 'Whitton & Nienow', 'Yawalkar','Zhu'])
plt.axis([0, 20, 0.1, 300])
plt.xlabel('v_gas [m/s]')
plt.ylabel('KLa [1/s]')
#plt.yscale('log')
```

There is quite a range. Smith and Warmoeskerken seems to be a real outlier, so will be ignored.

For following modelling, the fitted parameters of Whitton & Nienow will be used, both because they are close to the average of all curves, and because their parameter B = 1, meaning that it falls away as the exponent.

Hence, the formula to estimate the kLa of oxygen in a CSTR can be simplified to:

<div style="text-align: center">  $ k_La = 3.5 * v_{gs}$ </div>
