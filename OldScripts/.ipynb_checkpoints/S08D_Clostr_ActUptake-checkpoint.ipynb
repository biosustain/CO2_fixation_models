# Acetone producing Clostridium acetobutylicum

```python
Clostr = cobra.io.read_sbml_model(join(data_dir,'Models/iCac802.xml'))
```

## Anaerobic growth on Glucose

```python
medium = Clostr.medium
medium['Ex_1'] = 1000.0   # cpd00067, H+
medium['Ex_15'] = 1000.0  # cpd00013, NH3
medium['Ex_49'] = 1000.0  # cpd00009, Phosphate
medium['Ex_91'] = 10    # cpd00027, Glucose
medium['Ex_96'] = 0     # cpd00211, Butyrate
medium['Ex_97'] = 1000.0  # cpd00048, Sulfate
medium['Ex_101'] = 1000.0 # cpd00001, H2O
medium['Ex_103'] = 1000.0 # cpd00011, CO2
medium['Ex_104'] = 0 # cpd00029, Acetate
medium['Ex_112'] = 1000.0 # cpd00528, N2
Clostr.medium = medium
Clostr.medium

SolutionAc = Clostr.optimize()
Clostr.summary()
```
